```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def lwlr(x, y, k):
    m = len(x)
    y_pred = np.zeros(m)
    for i in range(m):
        weights = np.exp(-(x - x[i])**2 / (2 * k**2))
        W = np.diag(weights)
        X = np.vstack([np.ones(m), x]).T
        theta = np.linalg.inv(X.T @ W @ X) @ (X.T @ W @ y)
        y_pred[i] = np.dot([1, x[i]], theta)
    return y_pred

data = pd.read_csv('10-dataset.csv')
bill = data['total_bill'].values
tip = data['tip'].values

k = 0.5

y_pred = lwlr(bill, tip, k)
sort_idx = np.argsort(bill)
bill_sorted = bill[sort_idx]
y_pred_sorted = y_pred[sort_idx]

plt.scatter(bill, tip, color='green', label='Data points')
plt.plot(bill_sorted, y_pred_sorted, color='red', linewidth=2, label='Prediction')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
```
