```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets,preprocessing
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data,columns=iris.feature_names)
y = iris.target

X_scaled = preprocessing.StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=3).fit_predict(X)
gmm = GaussianMixture(n_components=3).fit_predict(X_scaled)

colormap = np.array(['red','lime','black'])
fig , axs = plt.subplots(2,2,figsize=(14,14))

axs[0,0].scatter(X.iloc[:,2],X.iloc[:,3],c=colormap[y] ,s=40)
axs[0,0].set_title("real")
axs[0,0].set_xlabel("petal length")
axs[0,0].set_ylabel("petal_width")

axs[0,1].scatter(X.iloc[:,2],X.iloc[:,3],c=colormap[kmeans] ,s=40)
axs[0,1].set_title("real")
axs[0,1].set_xlabel("petal length")
axs[0,1].set_ylabel("petal_width")

axs[1,0].scatter(X.iloc[:,2],X.iloc[:,3],c=colormap[gmm] ,s=40)
axs[1,0].set_title("real")
axs[1,0].set_xlabel("petal length")
axs[1,0].set_ylabel("petal_width")
```
