```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def kernel(point, xmat, k):
    m, n = np.shape(xmat)
    weights = np.mat(np.eye(m))
    for j in range(m):
        diff = point - xmat[j]
        weights[j, j] = np.exp(diff * diff.T / (-2.0 * k**2))
    return weights

def localWeight(point, xmat, ymat, k):
    wei = kernel(point, xmat, k)
    W = (xmat.T * (wei * xmat)).I * (xmat.T * (wei * ymat.T))
    return W

def localWeightRegression(xmat, ymat, k):
    m, n = np.shape(xmat)
    ypred = np.zeros(m)
    for i in range(m):
        ypred[i] = xmat[i] * localWeight(xmat[i], xmat, ymat, k)
    return ypred

# Load data points
data = pd.read_csv('10-dataset.csv')
bill = np.array(data.total_bill)
tip = np.array(data.tip)

# Preparing and adding 1 in bill
mbill = np.mat(bill)
mtip = np.mat(tip)
m = np.shape(mbill)[1]
one = np.mat(np.ones(m))
X = np.hstack((one.T, mbill.T))

# Set kernel bandwidth (k)
k = 0.5

# Perform local weighted regression
ypred = localWeightRegression(X, mtip, k)

# Sort the predictions
SortIndex = X[:, 1].argsort(0)
xsort = X[SortIndex][:, 0]

# Plot the data points and the regression line
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(bill, tip, color='green')
ax.plot(xsort[:, 1], ypred[SortIndex], color='red', linewidth=5)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Function to calculate weights
def kernel(point, xmat, k):
    m = np.shape(xmat)[0]
    weights = np.eye(m)
    for j in range(m):
        diff = point - xmat[j]
        weights[j, j] = np.exp(np.dot(diff, diff.T) / (-2.0 * k**2))
    return weights

# Function to calculate local weights
def localWeight(point, xmat, ymat, k):
    weights = kernel(point, xmat, k)
    W = np.linalg.inv(xmat.T @ weights @ xmat) @ xmat.T @ weights @ ymat
    return W

# Function for local weighted regression
def localWeightRegression(xmat, ymat, k):
    m = np.shape(xmat)[0]
    ypred = np.zeros(m)
    for i in range(m):
        ypred[i] = xmat[i] @ localWeight(xmat[i], xmat, ymat, k)
    return ypred

# Load data
data = pd.read_csv('10-dataset.csv')
bill = data['total_bill'].values
tip = data['tip'].values

# Prepare data by adding a column of ones
X = np.vstack((np.ones(len(bill)), bill)).T
y = tip

# Set the bandwidth parameter
k = 0.5

# Perform local weighted regression
ypred = localWeightRegression(X, y, k)

# Sort the values for plotting
sort_index = X[:, 1].argsort()
xsort = X[sort_index]

# Plot the results
plt.scatter(bill, tip, color='green', label='Data points')
plt.plot(xsort[:, 1], ypred[sort_index], color='red', linewidth=2, label='Prediction')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Calculate weights for a given point
def kernel(point, xmat, k):
    diff = xmat - point
    squared_diff = np.sum(diff ** 2, axis=1)
    weights = np.exp(-squared_diff / (2 * k ** 2))
    return np.diag(weights)

# Perform local weighted regression
def localWeightRegression(xmat, ymat, k):
    m = np.shape(xmat)[0]
    ypred = np.zeros(m)
    for i in range(m):
        weights = kernel(xmat[i], xmat, k)
        W = np.linalg.inv(xmat.T @ weights @ xmat) @ xmat.T @ weights @ ymat
        ypred[i] = xmat[i] @ W
    return ypred

# Load data
data = pd.read_csv('10-dataset.csv')
bill = data['total_bill'].values
tip = data['tip'].values

# Prepare data by adding a column of ones
X = np.vstack((np.ones(len(bill)), bill)).T
y = tip

# Set the bandwidth parameter
k = 0.5

# Perform local weighted regression
ypred = localWeightRegression(X, y, k)

# Sort the values for plotting
sort_index = X[:, 1].argsort()
xsort = X[sort_index]

# Plot the results
plt.scatter(bill, tip, color='green', label='Data points')
plt.plot(xsort[:, 1], ypred[sort_index], color='red', linewidth=2, label='Prediction')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Local weighted regression
def lwlr(X, y, k):
    m = len(X)
    y_pred = np.zeros(m)
    for i in range(m):
        W = np.diag(np.exp(-np.sum((X[i] - X)**2, axis=1) / (2 * k**2)))
        theta = np.linalg.inv(X.T @ W @ X) @ X.T @ W @ y
        y_pred[i] = X[i] @ theta
    return y_pred

# Load data
data = pd.read_csv('10-dataset.csv')
bill, tip = data['total_bill'].values, data['tip'].values

# Prepare data
X = np.vstack((np.ones(len(bill)), bill)).T
k = 0.5

# Predict and sort
ypred = lwlr(X, tip, k)
sort_idx = bill.argsort()
bill_sorted, ypred_sorted = bill[sort_idx], ypred[sort_idx]

# Plot
plt.scatter(bill, tip, color='green', label='Data points')
plt.plot(bill_sorted, ypred_sorted, color='red', linewidth=1, label='Prediction')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Local weighted regression
def lwlr(x, y, k):
    m = len(x)
    y_pred = np.zeros(m)
    for i in range(m):
        weights = np.exp(-(x - x[i])**2 / (2 * k**2))
        W = np.diag(weights)
        X = np.vstack([np.ones(m), x]).T
        theta = np.linalg.inv(X.T @ W @ X) @ (X.T @ W @ y)
        y_pred[i] = np.dot([1, x[i]], theta)
    return y_pred

# Load data
data = pd.read_csv('10-dataset.csv')
bill = data['total_bill'].values
tip = data['tip'].values

# Prepare data
k = 0.5

# Predict and sort
y_pred = lwlr(bill, tip, k)
sort_idx = np.argsort(bill)
bill_sorted = bill[sort_idx]
y_pred_sorted = y_pred[sort_idx]

# Plot
plt.scatter(bill, tip, color='green', label='Data points')
plt.plot(bill_sorted, y_pred_sorted, color='red', linewidth=2, label='Prediction')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()
plt.show()
```
